```python
store_sales = (
    df.groupby("Store")["Weekly_Sales"]
      .sum()
      .sort_values(ascending=False)
)

print(store_sales)
```

```python
import matplotlib.pyplot as plt

top10 = store_sales.head(10)

plt.figure(figsize=(10,5))
top10.plot(kind='bar')
plt.title('Top 10 Stores by Total Sales')
plt.ylabel('Total Sales')
plt.show()
```

```python
avg_sales = (
    df.groupby("Store")["Weekly_Sales"]
      .mean()
      .sort_values(ascending=False)
)

print(avg_sales.head(10))
print(avg_sales.tail(10))
```

```python
holiday_sales = (
    df.groupby("Holiday_Flag")["Weekly_Sales"]
      .mean()
)

print(holiday_sales)
```

```python
holiday_lift = (
    (1122888 - 1041256)
    / 1041256
) * 100

print(holiday_lift)
```

```python
holiday_lift = (
    (holiday_sales[1] - holiday_sales[0])
    / holiday_sales[0]
) * 100

print(f"Holiday Sales Lift: {holiday_lift:.2f}%")
```

```python
corr = df.corr(numeric_only=True)

corr["Weekly_Sales"].sort_values(ascending=False)
```

```python
total_sales = df["Weekly_Sales"].sum()

top10_sales = (
    df.groupby("Store")["Weekly_Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
      .sum()
)

print(f"Total Sales: ${total_sales:,.0f}")
print(f"Top 10 Stores Sales: ${top10_sales:,.0f}")
print(f"Revenue Share: {(top10_sales/total_sales)*100:.2f}%")
```

```python
import matplotlib.pyplot as plt

store_sales = (
    df.groupby("Store")["Weekly_Sales"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))
store_sales.plot(kind="bar")

plt.title("Total Sales by Store")
plt.xlabel("Store")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

```python
total_sales = df["Weekly_Sales"].sum()

store_sales = (
    df.groupby("Store")["Weekly_Sales"]
      .sum()
      .sort_values(ascending=False)
)

top10_sales = store_sales.head(10).sum()

revenue_share = (top10_sales / total_sales) * 100

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Top 10 Store Sales: ${top10_sales:,.2f}")
print(f"Revenue Share: {revenue_share:.2f}%")
```

```python
store_stats = (
    df.groupby("Store")["Weekly_Sales"]
      .agg(["mean", "std", "min", "max"])
      .sort_values("mean", ascending=False)
)

store_stats.head(10)
```

```python
store_stats["cv"] = store_stats["std"] / store_stats["mean"]

store_stats.sort_values("cv").head(10)
```

```python
store_stats.sort_values("cv", ascending=False).head(10)
```

```python
holiday_store = (
    df.groupby(["Store", "Holiday_Flag"])["Weekly_Sales"]
      .mean()
      .unstack()
)

holiday_store["Holiday_Lift_%"] = (
    (holiday_store[1] - holiday_store[0])
    / holiday_store[0]
) * 100

holiday_store.sort_values(
    "Holiday_Lift_%",
    ascending=False
).head(10)
```

```python
top_stores = (
    df.groupby("Store")["Weekly_Sales"]
      .agg(["mean"])
      .sort_values("mean", ascending=False)
      .head(10)
)

bottom_stores = (
    df.groupby("Store")["Weekly_Sales"]
      .agg(["mean"])
      .sort_values("mean", ascending=True)
      .head(10)
)

print(top_stores)
print(bottom_stores)
```
